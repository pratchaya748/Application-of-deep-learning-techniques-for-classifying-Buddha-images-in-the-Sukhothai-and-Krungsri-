```python
from PIL import Image
import os
import PIL
import glob
```

```python
def manageFileDir(filepath):
    dirname, basename = os.path.split(filepath)
    basename_noext = os.path.splitext(os.path.basename(filepath))[0]
    return dirname, basename, basename_noext
```

```python
def resizeImage (imgdir, destdir, fixed_height):
    
    filelist = glob.glob(imgdir)
    for filepath in filelist:

        dirname, basename, basename_noext = manageFileDir(filepath)

        image = Image.open(filepath)
        x, y  = image.size
#         height_percent = (fixed_height / float(image.size[1]))
#         width_size = int((float(image.size[0]) * float(height_percent)))
        width_size = fixed_height
        image = image.resize((width_size, fixed_height), PIL.Image.NEAREST)
        filesave = destdir+ basename_noext  + '.jpg'
        image.save(filesave)
        print(filesave+'...saving...done!')
    
 
```

```python
imgdir = 'กลุ่ม-3-Buddha-Images/Ayutthaya/*.jpg'
destdir= 'resized200x200/Ayutthaya/'
```

```python
fixed_height = 200
resizeImage(imgdir, destdir, fixed_height)
```
